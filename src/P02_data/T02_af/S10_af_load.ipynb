```python
import csv
import os
import pickle
import re
from pathlib import Path
import pandas as pd
```

```python
CURRENT_DIR = Path.cwd()
PARENT_DIR = CURRENT_DIR.parent
DATA_DIR = PARENT_DIR / "T00_data" / "axial_force" / "post_welding"
print(CURRENT_DIR)
print(DATA_DIR)
```

```python
# Extract the number from the filename
def extract_number(filename):
    # Use regular expression to find the number in the filename
    match = re.search(r"(\d+)", filename)
    if match:
        return int(match.group(1))
    return None


# Function to process the raw data into a structured list of 2-item lists
def process_csv_lines(file):
    processed_data = []
    # Use the csv reader to handle quotes and commas properly
    reader = csv.reader(file, delimiter=",")
    for row in reader:
        # A valid row should have at least 2 elements (key and value)
        if len(row) >= 2:
            # Join any extra elements back into the second column (the 'Value')
            key = row[0]
            value = ", ".join(row[1:])
            processed_data.append(
                [key, value.strip()]
            )  # Strip any leading/trailing space from value
    return processed_data
```

```python
# List all files in the data directory
files = os.listdir(DATA_DIR)
```

## Load metadata from CSV file

```python
df_list = []
for file in files:
    filepath = DATA_DIR / file
    with open(filepath, "r", encoding="utf-8") as f:
        data = process_csv_lines(f)
    sample_no = extract_number(file)
    df = pd.DataFrame(data)
    df.rename(columns={0: "Parameter", 1: "Value"}, inplace=True)
    df["Parameter"] = df["Parameter"].str.strip().str.replace(":", "")
    df.set_index("Parameter", inplace=True)
    df.index.name = None
    df.columns = ["Value"]
    _meta = df["Value"].iloc[:18]
    _meta["filename"] = file
    _meta["sample_no"] = sample_no
    df_list.append(_meta)

metas = pd.DataFrame(df_list)
metas.reset_index(drop=True, inplace=True)
metas.set_index("sample_no", inplace=True)
metas.sort_index(inplace=True)
metas
```

## Load time series from csv file

```python
df_list = []
for file in files:
    filepath = DATA_DIR / file
    sample_no = extract_number(file)
    afs = pd.read_csv(filepath, skiprows=18)
    afs = afs.iloc[1:, :]  # Skip the first row after the header because it contains units
    # Convert columns to numeric, coerce errors to NaN
    afs["Time"] = pd.to_numeric(afs["Time"], errors="coerce")
    afs["Fx"] = pd.to_numeric(afs["Fx"], errors="coerce")
    afs["Fy"] = pd.to_numeric(afs["Fy"], errors="coerce")
    afs["Fz"] = pd.to_numeric(afs["Fz"], errors="coerce")
    afs["Mz"] = pd.to_numeric(afs["Mz"], errors="coerce")
    afs["sample_no"] = sample_no
    # af.columns = af.columns.str.strip().str.lower()
    df_list.append(afs)

afs = pd.concat(df_list)
afs.reset_index(drop=True, inplace=True)
afs
```

## Save files

```python
metas.to_excel("S10_af_meta.xlsx", index=True)

with open("S10_af_time_series.pkl", "wb") as f:
    pickle.dump(dict(data=afs), f)
```
