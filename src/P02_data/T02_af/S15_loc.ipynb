## Calculate position for meausred points


```python
distance_buffer_start = 32 - 12  # mm
distance_between_points = 15  # mm

position = distance_buffer_start
locs_data = []
for idx in range(7):
    locs_data.append(dict(loc=idx + 1, position=position))
    position += distance_between_points

locs = pd.DataFrame(locs_data)
display(locs)
locs.to_excel("S15_loc_positions.xlsx", index=False)
```

## Function


```python
def af_loc(
    sample_no,
    distance_buffer_start=distance_buffer_start,
    distance_between_points=distance_between_points,
    file_prefix="S15",
):
    # sample_no = 1
    # Extract relevant data
    _weld_speed = metas.loc[metas["sample_no"] == sample_no, "W"].values[0]  # mm/min
    weld_speed = _weld_speed / 60  # Convert to mm/s
    af = afs.loc[afs["sample_no"] == sample_no].reset_index(drop=True)
    af.set_index("Time", inplace=True)
    weld_time_end = weld_end.loc[
        weld_end["sample_no"] == sample_no, "weld_time_end"
    ].values[0]
    dwell_time_end = dwell.loc[
        dwell["sample_no"] == sample_no, "dwell_time_end"
    ].values[0]
    weld_time_start = dwell_time_end

    # distance_buffer_start = 32 - 12  # mm  # Check with Parn to confirm this value.
    # distance_between_points = 15  # mm
    time_buffer_start = distance_buffer_start / weld_speed  # s
    time_between_points = distance_between_points / weld_speed  # s
    loc_times = [weld_time_start + time_buffer_start]
    for i in range(6):
        loc_times.append(loc_times[-1] + time_between_points)
    print(loc_times)
    loc_times_idxs = af.index.get_indexer(loc_times, method="nearest")
    af_mean = af.rolling(window=30).mean()
    loc_fzs = af_mean["Fz"].iloc[loc_times_idxs].values

    fig, ax = plt.subplots(1, 1, figsize=(12, 6))
    ax.plot(af.index, af["Fz"])
    ax.plot(af_mean.index, af_mean["Fz"], color="orange", label="Fz (Smoothed)")
    ax.axvline(weld_time_start, color="red", linestyle="--", label="Dwell End")
    ax.plot(af.index[loc_times_idxs], loc_fzs, "o", label="LOC Points", color="red")
    for i, (loc, loc_idx) in enumerate(zip(loc_times, loc_times_idxs)):
        ax.axvline(loc, color="green", linestyle="--")
        ax.text(
            loc,
            af["Fz"].iloc[loc_idx] + af["Fz"].max() * 0.2,
            f"Loc-{i + 1}",
            verticalalignment="center",
            color="green",
            clip_on=True,
        )
    ax.axvline(weld_time_end, color="blue", linestyle="--", label="Weld End")
    ax.set_xlabel("Time (s)")
    ax.set_ylabel("Fz (kN)")

    fig.suptitle(f"Sample No: {sample_no} | Weld Speed: {_weld_speed} mm/min")
    fig.tight_layout()
    fig.savefig(
        f"{OUT_DIR}/{file_prefix}_sample_{sample_no}.png", dpi=300, bbox_inches="tight"
    )
    plt.close(fig)

    data_dict = {}
    for i, (loc, loc_idx, fz) in enumerate(zip(loc_times, loc_times_idxs, loc_fzs)):
        data_dict[f"LOC{i + 1}_time"] = loc
        data_dict[f"LOC{i + 1}_Fz"] = fz
        data_dict[f"LOC{i + 1}_idx"] = loc_idx
    return data_dict
```

```python
_params = []
for sample_no in metas["sample_no"].unique():
    _params.append(
        dict(
            sample_no=sample_no,
        )
    )
params = pd.DataFrame.from_dict(_params)
display(params)
```

```python
reses = []
for param in params.to_dict(orient="records")[:]:
    print(param)
    res = af_loc(**param)
    res = {**param, **res}
    reses.append(res)
    print("----------")

df_loc = pd.DataFrame.from_dict(reses)
```
