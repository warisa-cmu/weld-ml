```python
import os
import pickle
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
with open("S10_af_time_series.pkl", "rb") as f:
    data = pickle.load(f)

afs = data["data"]
infos = pd.read_excel("S16_af_info.xlsx")
```

```python
infos
```

```python
infos.columns
```

```python
datas = []
for sample_no in infos["sample_no"].unique():
    # Extract relevant data
    af = afs.loc[afs["sample_no"] == sample_no].reset_index(drop=True)
    af.set_index("Time", inplace=True)
    info = infos[infos["sample_no"] == sample_no].iloc[0]

    # Extract Fz in different phases
    af_contact = af.loc[info["contact_time_start"] : info["dwell_time_start"]]
    af_dwell = af.loc[info["dwell_time_start"] : info["dwell_time_end"]]
    af_weld = af.loc[info["dwell_time_end"] : info["weld_time_end"]]
    
    datas.append({
        "sample_no": sample_no,
        "af_contact": af_contact,
        "af_dwell": af_dwell,
        "af_weld": af_weld,
    })
    
af_extract = pd.DataFrame(datas)
```
