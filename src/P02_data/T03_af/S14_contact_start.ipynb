```python
# Detect the beginning of the time series as the start of contact
sample_no = 5
shoot_idx_chosen = 0
shoot_threshold = 0.5  # Absolute threshold for sudden shoot

# Extract relevant data
af = afs.loc[afs["sample_no"] == sample_no].reset_index(drop=True)
af.set_index("Time", inplace=True)
dwell_time_start = dwell.loc[dwell["sample_no"] == sample_no, "dwell_time_start"].values[0]

# Smooth the axial force signal 
afsm = af['Fz'].rolling(window=20, min_periods=1).mean()

# Find the beginning of contact based on sudden shot in axial force
fzDiff = afsm.diff()  # First difference

# Zero value during the 10 points of the initial period to avoid edge effects
fzDiff.iloc[: 10] = 0

z = (fzDiff - fzDiff.mean()) / fzDiff.std()  # Normalize (z-score)
filt_shoot = z > shoot_threshold  # Boolean filter for sudden shoot
shoot_idxs = filt_shoot.reset_index()[
    filt_shoot.values
].index  # Get the numerical indices of the drops

if (
    shoot_idx_chosen < 10
):  # If the number of small, I want to choose from the list of detected shoots.
    contact_time_start_idx = shoot_idxs[
        shoot_idx_chosen
    ]  # Choose the desired shoot index
else:  # If the number is large, I want to choose the index directly.
    contact_time_start_idx = shoot_idx_chosen  # Choose the desired shoot index
contact_time_start = af.index[
    contact_time_start_idx
]  # Get the time of the contact start

# Calculate contact duration
contact_time_actual = dwell_time_start - contact_time_start

# Calculate the pre contact level
pre_contact_level = (
    af["Fz"].iloc[:contact_time_start_idx].mean()
)  # Mean axial force before contact

# Plotting
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 8))
ax1.plot(af.index, af["Fz"], label="Axial Force (Fz)")
ax1.axvline(
    contact_time_start, color="red", linestyle="--", label="Contact Start Time"
)
ax1.set_ylabel("Axial Force (Fz)")

# Write shoot indices on the plot
xlim = [
    contact_time_start - af.index.max() / 10,
    contact_time_start + af.index.max() / 10,
]
ax2.plot(af.index, af["Fz"], label="Axial Force (Fz)")
ax2.axvline(
    contact_time_start, color="red", linestyle="--", label="Contact Start Time"
)
ax2.set_ylabel("Axial Force (Fz)")
ax2.set_xlim(xlim)

ax3.plot(af.index, z, label="First Difference of Fz", color="orange")
ax3.axhline(shoot_threshold, color="red", linestyle="--", label="Shoot Threshold")
ax3.set_ylabel("First Difference of Fz (Z-score)")
ax3.set_xlabel("Time (s)")
# ax3.set_xlim(xlim)

```

```python
def shoot_detection(sample_no, shoot_idx_chosen, shoot_threshold, file_prefix="S14"):
    # Detect the beginning of the time series as the start of contact
    # sample_no = 1
    # shoot_idx_chosen = -1
    # shoot_threshold = -4  # Absolute threshold for sudden shoot detection

    # Extract relevant data
    af = afs.loc[afs["sample_no"] == sample_no].reset_index(drop=True)
    af.set_index("Time", inplace=True)
    dwell_time_start = dwell.loc[dwell["sample_no"] == sample_no, "dwell_time_start"].values[0]

    # Smooth the axial force signal 
    afsm = af['Fz'].rolling(window=20, min_periods=1).mean()

    # Find the beginning of contact based on sudden shot in axial force
    fzDiff = afsm.diff()  # First difference
    
    # Zero value during the 10 points of the initial period to avoid edge effects
    fzDiff.iloc[: 10] = 0

    z = (fzDiff - fzDiff.mean()) / fzDiff.std()  # Normalize (z-score)
    filt_shoot = z > shoot_threshold  # Boolean filter for sudden shoot
    shoot_idxs = filt_shoot.reset_index()[
        filt_shoot.values
    ].index  # Get the numerical indices of the drops

    if (
        shoot_idx_chosen < 10
    ):  # If the number of small, I want to choose from the list of detected shoots.
        contact_time_start_idx = shoot_idxs[
            shoot_idx_chosen
        ]  # Choose the desired shoot index
    else:  # If the number is large, I want to choose the index directly.
        contact_time_start_idx = shoot_idx_chosen  # Choose the desired shoot index
    contact_time_start = af.index[
        contact_time_start_idx
    ]  # Get the time of the contact start

    # Calculate contact duration
    contact_time_actual = dwell_time_start - contact_time_start

    # Calculate the pre contact level
    pre_contact_level = (
        af["Fz"].iloc[:contact_time_start_idx].mean()
    )  # Mean axial force before contact

    # Plotting
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 8))
    ax1.plot(af.index, af["Fz"], label="Axial Force (Fz)")
    ax1.axvline(
        contact_time_start, color="red", linestyle="--", label="Contact Start Time"
    )
    ax1.set_ylabel("Axial Force (Fz)")

    # Write shoot indices on the plot
    xlim = [
        contact_time_start - af.index.max() / 10,
        contact_time_start + af.index.max() / 10,
    ]
    ax2.plot(af.index, af["Fz"], label="Axial Force (Fz)")
    ax2.axvline(
        contact_time_start, color="red", linestyle="--", label="Contact Start Time"
    )
    ax2.set_ylabel("Axial Force (Fz)")
    ax2.set_xlim(xlim)

    ax3.plot(af.index, z, label="First Difference of Fz", color="orange")
    ax3.axhline(shoot_threshold, color="red", linestyle="--", label="Shoot Threshold")
    ax3.set_ylabel("First Difference of Fz (Z-score)")
    ax3.set_xlabel("Time (s)")
    # ax3.set_xlim(xlim)

    fig.suptitle(
        f"Sample No: {sample_no} - Weld End Time Detection, Index: {contact_time_start_idx}, Time: {contact_time_start:.2f}s"
    )
    fig.savefig(
        f"{OUT_DIR}/{file_prefix}_S{str(sample_no).zfill(2)}.png",
        dpi=300,
        bbox_inches="tight",
    )
    plt.close(fig)
    return dict(
        contact_time_start_idx=contact_time_start_idx,
        contact_time_start=contact_time_start,
        pre_contact_level=pre_contact_level,
        contact_time_actual=contact_time_actual,
    )
```

```python
default_shoot_threshold = 0.5
default_shoot_idx_chosen = 0

_params = []
for sample_no in metas["sample_no"].unique():
    _params.append(
        dict(
            sample_no=sample_no,
            shoot_idx_chosen=default_shoot_idx_chosen,
            shoot_threshold=default_shoot_threshold,
        )
    )


params = pd.DataFrame.from_dict(_params)
params
```

```python
reses = []
for param in params.to_dict(orient="records")[:]:
    print(param)
    res = shoot_detection(**param)
    res = {**param, **res}
    reses.append(res)

dfwe = pd.DataFrame.from_dict(reses)
```
