```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
from pathlib import Path
```

```python
CURRENT_DIR = Path.cwd()
PARENT_DIR = CURRENT_DIR.parent
DATA_DIR = PARENT_DIR / "T04_preprocess"
print(CURRENT_DIR)
print(DATA_DIR)
```

```python
filepath = os.path.join(DATA_DIR, "S10_center_location_7_individual.xlsx")
dfm = pd.read_excel(filepath)
dfm
```

```python
# Plot boxenplot for sigma_x_post by location, filtered for Center section
filt = dfm["section"].isin(["Center"])
fig, ax = plt.subplots(figsize=(10, 6))
sns.boxenplot(dfm[filt], x="location", y="sigma_x_post", hue="measurement", ax=ax)
ax.set_title("Post-weld Residual Stress Distribution by Location and Center Section")
```

```python
# Plot boxenplot for FWHM_post by location, filtered for Center section
filt = dfm["section"].isin(["Center"])
fig, ax = plt.subplots(figsize=(10, 6))
sns.boxenplot(dfm[filt], x="location", y="FWHM_post", hue="measurement", ax=ax)
ax.set_title("Post-weld FWHM Distribution by Location and Center Section")
```
