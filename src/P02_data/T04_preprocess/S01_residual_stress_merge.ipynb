```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
from pathlib import Path
```

```python
CURRENT_DIR = Path.cwd()
PARENT_DIR = CURRENT_DIR.parent
DATA_DIR = PARENT_DIR / "T02_stress"
print(CURRENT_DIR)
print(DATA_DIR)
```

```python
filepath_pre = os.path.join(DATA_DIR, "S01_residual_stress_pre_weld.xlsx")
filepath_post = os.path.join(DATA_DIR, "S02_residual_stress_post_weld_first.xlsx")
pre = pd.read_excel(filepath_pre).rename(columns={"Sigma(x)": "sigma_x_pre", "FWHM": "FWHM_pre", "idx_excel": "idx_excel_pre"})
post = pd.read_excel(filepath_post).rename(columns={"Sigma(x)": "sigma_x_post", "FWHM": "FWHM_post", "idx_excel": "idx_excel_post"})
```
