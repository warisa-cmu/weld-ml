```python
import pickle
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
CURRENT_DIR = Path.cwd()
PARENT_DIR = CURRENT_DIR.parent
AF_DATA_DIR = PARENT_DIR / "T03_af"
print(AF_DATA_DIR)
```

```python
# Load AF data
pickle_filepath = AF_DATA_DIR / "S17_af_extract.pkl"
with open(pickle_filepath, "rb") as handle:
    data = pickle.load(handle)
afs = data["data"]
```

```python
afs
```

```python
def reample(_df):
    df = _df.copy()
    # Assume the index is in seconds, convert to milliseconds
    df.index = pd.to_timedelta(df.index.values * 1000, unit="ms")
    # Downsample Af to 0.1 Hz
    df = df.resample("100ms").mean().interpolate()
    # Convert index back to seconds
    df.index = df.index.total_seconds()
    return df
```

```python
data_arr = []
for sample_no in afs["sample_no"].unique():
    # print(sample_no)
    fz_dwell = afs.loc[afs["sample_no"] == sample_no, "af_dwell"].values[0]["Fz"]
    fz_weld = afs.loc[afs["sample_no"] == sample_no, "af_weld"].values[0]["Fz"]
    fz_dwell_ds = reample(fz_dwell)
    fz_weld_ds = reample(fz_weld)

    # fig, ax = plt.subplots(1, 1, figsize=(10, 6), sharex=True)
    # ax.plot(fz_dwell.index, fz_dwell.values, label="Original")
    # ax.plot(fz_dwell_ds.index, fz_dwell_ds.values, label="Downsampled", linestyle="--")

    data = {
        "sample_no": sample_no,
        "fz_dwell": fz_dwell_ds,
        "fz_weld": fz_weld_ds,
    }
    
    print(f"Sample No: {sample_no}, Original Length: {len(fz_dwell)}, Downsampled Length: {len(fz_dwell_ds)}")
    print(f"Sample No: {sample_no}, Original Length: {len(fz_weld)}, Downsampled Length: {len(fz_weld_ds)}")

    data_arr.append(data)
    
downsampled_afs = pd.DataFrame(data_arr)
downsampled_afs
```

```python
with open("S02_downsampled.pkl", "wb") as f:
    pickle.dump(dict(data=downsampled_afs), f)
```
