```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
from pathlib import Path
```

```python
CURRENT_DIR = Path.cwd()
PARENT_DIR = CURRENT_DIR.parent
DATA_DIR = PARENT_DIR / "T04_preprocess"
print(CURRENT_DIR)
print(DATA_DIR)
```

```python
filepath = os.path.join(DATA_DIR, "./S02/S02_residual_stress_imputed_20260104_114536.xlsx")
dfm = pd.read_excel(filepath)
dfm
```

```python
# Filter only the center section
filt = dfm["section"] == "Center"
dfm_center = dfm[filt]
dfm_center
```

```python
g = sns.boxenplot(data=dfm_center, x="location", y="diff_sigma_x")
g.axes.set_title("Residual Stress vs Location by R")
```
