```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
from pathlib import Path
```

```python
CURRENT_DIR = Path.cwd()
```

```python
filepath = os.path.join(CURRENT_DIR, "./S04/S04_residual_stress_imputed_20260104_114536.xlsx")
dfm = pd.read_excel(filepath)
dfm
```

```python
g = sns.boxenplot(data=dfm, x="section", y="diff_sigma_x", hue="R", palette="Set2")
g.axes.set_title("Residual Stress vs Section by R")
```

```python
g = sns.boxenplot(data=dfm, x="section", y="diff_sigma_x", hue="W", palette="Set2")
g.axes.set_title("Residual Stress vs Section by W")
```

```python
g = sns.boxenplot(data=dfm, x="section", y="diff_sigma_x", hue="D", palette="Set2")
g.axes.set_title("Residual Stress vs Section by D")
```

```python
g = sns.FacetGrid(data=dfm, col="location", col_wrap=4, height=4, sharey=True)
g.map_dataframe(sns.boxenplot, x="section", y="diff_sigma_x", hue="R", palette="Set2")
g.figure.suptitle("Residual Stress vs Section by Location and R", y=1.02)
g.add_legend()
```

```python
g = sns.FacetGrid(data=dfm, col="location", col_wrap=4, height=4, sharey=True)
g.map_dataframe(sns.boxenplot, x="section", y="diff_sigma_x", hue="W", palette="Set2")
g.figure.suptitle("Residual Stress vs Section by Location and W", y=1.02)
g.add_legend()
```
