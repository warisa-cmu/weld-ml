```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
from pathlib import Path
```

```python
CURRENT_DIR = Path.cwd()
PARENT_DIR = CURRENT_DIR.parent
DATA_DIR = PARENT_DIR / "T00_data"
print(CURRENT_DIR)
print(DATA_DIR)
```

```python
filepath_AA6061 = os.path.join(
    DATA_DIR, "residual_stress", "pre_welding", "Pre_RS (AA6061-T6).xlsx"
)
filepath_AA5052 = os.path.join(
    DATA_DIR, "residual_stress", "pre_welding", "Pre_RS (AA5052-H32).xlsx"
)
pre_AA6061 = pd.read_excel(filepath_AA6061)
pre_AA5052 = pd.read_excel(filepath_AA5052)
```

```python
pre_AA6061
```

```python
# Split grade and sample no and location


def split_data_1(el):
    sps = el.split("_")
    return sps[0], sps[1], sps[2]


pre_AA6061[["grade", "sample_no", "location"]] = pre_AA6061["Comment 1"].apply(
    lambda x: pd.Series(split_data_1(x))
)
pre_AA5052[["grade", "sample_no", "location"]] = pre_AA5052["Comment 1"].apply(
    lambda x: pd.Series(split_data_1(x))
)
```

```python
# Split data 2


def split_data_2(el):
    sps = el.split(",")
    r = int(sps[0][1:])
    w = int(sps[1][1:])
    d = int(sps[2][1:])
    return r, w, d


pre_AA6061[["R", "W", "D"]] = pre_AA6061["Comment 1"].apply(
    lambda x: pd.Series(split_data_1(x))
)
pre_AA5052[["R", "W", "D"]] = pre_AA5052["Comment 1"].apply(
    lambda x: pd.Series(split_data_1(x))
)
```
