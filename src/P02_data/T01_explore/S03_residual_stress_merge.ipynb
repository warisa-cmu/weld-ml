```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
from pathlib import Path
```

```python
CURRENT_DIR = Path.cwd()
```

```python
filepath_pre = os.path.join(CURRENT_DIR, "S01_residual_stress_pre_weld.xlsx")
filepath_post = os.path.join(CURRENT_DIR, "S02_residual_stress_post_weld.xlsx")
pre = pd.read_excel(filepath_pre).rename(columns={"Sigma(x)": "sigma_x_pre", "FWHM": "FWHM_pre", "idx_excel": "idx_excel_pre"})
post = pd.read_excel(filepath_post).rename(columns={"Sigma(x)": "sigma_x_post", "FWHM": "FWHM_post", "idx_excel": "idx_excel_post"})
```

```python
dfm = post.merge(
    pre, how="left", on=["location", "section", "sample_no", "R", "W", "D"]
).fillna(0)
dfm
```

```python
dfm['diff_sigma_x'] = dfm['sigma_x_post'] - dfm['sigma_x_pre']
```

```python
dfm.to_excel("S03_residual_stress_merge.xlsx", index=False)
```
