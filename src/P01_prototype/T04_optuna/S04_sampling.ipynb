```python
import optuna
import random
```

```python
def objective(trial):
    # Categorical parameter
    alpha = trial.suggest_categorical("alpha", ["Catagory1", "Catagory2", "Catagory3"])

    # Integer alphaeter
    # alpha = trial.suggest_int("alpha", 1, 4)

    # Integer alphaeter (log)
    # Mechanism: Optuna samples a real number uniformly from [log(low), log(high)], exponentiates it, 
    # and then rounds the result to the nearest integer within the specified range. 
    # alpha = trial.suggest_int("alpha", 32, 512, log=True) 

    # Integer alphaeter (discretized)
    # alpha = trial.suggest_int("alpha", 10, 100, step=5)

    # Floating point alphaeter
    # alpha = trial.suggest_float("alpha", 0.0, 1.0)

    # Floating point alphaeter (log)
    # alpha = trial.suggest_float("alpha", 1e-5, 1e-2, log=True)

    # Floating point alphaeter (discretized)
    # alpha = trial.suggest_float("alpha", 0.0, 1.0, step=0.1)

    return random.random()


optuna.logging.set_verbosity(optuna.logging.WARNING)
study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=50)
df = study.trials_dataframe()
# display(df)
df["params_alpha"].values
```
