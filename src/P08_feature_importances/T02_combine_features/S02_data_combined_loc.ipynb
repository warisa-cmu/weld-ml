```python
import re
from pathlib import Path
import pandas as pd
```

```python
CURRENT_DIR = Path.cwd()
DATA_AF_DIR = CURRENT_DIR.parent.parent / "P02_data" / "T07_af_all"
DATA_COMBINED_DIR = (
    CURRENT_DIR.parent.parent / "P03_run_diff_sigma" / "T02_combine_features"
)
DATA_FEATURES_DIR = CURRENT_DIR.parent / "T01_af_features"
print(DATA_AF_DIR)
print(DATA_COMBINED_DIR)
print(DATA_FEATURES_DIR)
```

```python
# Load selected features
features = pd.read_excel(DATA_FEATURES_DIR / "S05_af_selected_features.xlsx")
features
```

```python
# Load location info
locs = pd.read_excel(DATA_AF_DIR / "S01_info_locs.xlsx").rename(
    columns={
        "Location": "location",
        "Fx": "Fx_location",
        "Fy": "Fy_location",
        "Fz": "Fz_location",
        "Mz": "Mz_location",
    }
)
locs
```

```python
# Load combined data
_combined = pd.read_excel(DATA_COMBINED_DIR / "s02_data_combined_loc.xlsx")

# Only keep relevant columns
cols_keep = [
    "sample_no",
    "R",
    "W",
    "D",
    "position",
    "location",
    "stress_value_5052",
    "stress_value_6061",
    "stress_value_center",
]
combined1 = _combined[cols_keep]
combined1
```

```python
combined2 = combined1.merge(locs, on=["sample_no", "location"])
combined2
```

```python
combined3 = combined2.merge(features, on=["sample_no"])
combined3
```

```python
colsY = ["stress_value_5052", "stress_value_6061", "stress_value_center"]
colsX = [c for c in combined3.columns if c not in colsY]
combined4 = combined3[colsX + colsY]
combined4
```

```python
combined4.to_excel("S02_data_combined_loc.xlsx", index=False)
```
