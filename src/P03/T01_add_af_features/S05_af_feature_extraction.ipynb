```python
import os
import pickle
import re
from pathlib import Path
import pandas as pd
```

```python
# Load features
features_dwell = pd.read_excel("S02_af_features.xlsx", sheet_name="af_dwell")
features_weld = pd.read_excel("S02_af_features.xlsx", sheet_name="af_weld")

# Load features relevance
select_dwell = pd.read_excel("S04_af_feature_selection.xlsx", sheet_name="af_dwell")
select_weld = pd.read_excel("S04_af_feature_selection.xlsx", sheet_name="af_weld")
```

```python
cols_dwell = select_dwell["feature"].tolist()
cols_dwell = ["sample_no", *cols_dwell]
cols_weld = select_weld["feature"].tolist()
cols_weld = ["sample_no", *cols_weld]
```

```python
fe_dwell = features_dwell[cols_dwell]
cols_mod = ["dwell_" + col for col in fe_dwell.columns if col != "sample_no"]
fe_dwell.columns = ["sample_no", *cols_mod]
display(fe_dwell)
```

```python
fe_weld = features_weld[cols_weld]
cols_mod = ["weld_" + col for col in fe_weld.columns if col != "sample_no"]
fe_weld.columns = ["sample_no", *cols_mod]
display(fe_weld)
```

```python
dfm = fe_dwell.merge(fe_weld, on="sample_no")
dfm
```

```python
dfm.to_excel("S05_af_features_extraction.xlsx", index=False)
```
